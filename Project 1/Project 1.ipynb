```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print (df)
```

# Gender Distribution 

```python
plt.figure(figsize = [6,4.5])
ax = sns.countplot(data = df , x = "Gender") 
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

##### From the above data we have analysed that the number of females in the data are more than the males

```python
gb = df.groupby("ParentEduc").agg({"MathScore" : "mean" ,"ReadingScore" :"mean" ,"WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(data = gb,annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

##### From the above chart we have concluded that the education of parents have a good impact on their children

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" : "mean" ,
                                             "ReadingScore" :"mean" ,"WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = [6,4])
sns.heatmap(data = gb1,annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

##### from the above data we have concluded that parent marital status have no or negligible impact on the score of the students

# Ethnic Group Distribution

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

label = ["group A", "group B", "group C", "group D", "group E"]

group = [groupA["EthnicGroup"] ,groupB["EthnicGroup"] ,groupC["EthnicGroup"] 
         ,groupD["EthnicGroup"] ,groupE["EthnicGroup"]]

plt.pie(group, autopct = "%.2f%%", labels = label, shadow = True) 
plt.title("Distribution Ethnic Group")
plt.show()
```

```python
a = sns.countplot(data = df , x = "EthnicGroup")
plt.bar_label(a.containers[0])
plt.show()
```

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore" : "mean" ,"ReadingScore" :"mean" ,"WritingScore":"mean"})
print(gb2)
```

```python
sns.heatmap(data = gb2,annot = True)
plt.title("Relationship between sports and Student's Score")
plt.show()
```

##### From the above data we have concluded that students who practice sports daily have scored good
