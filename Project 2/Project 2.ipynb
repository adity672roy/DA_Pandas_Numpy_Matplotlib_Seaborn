```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("HR.csv")
df
```

```python
df =df.drop_duplicates()
df
```

```python
df["Department"] = df["Department"].fillna("Analyst")
df
```

```python
age_mean = df["Age"].mean()
age = int(age_mean) 
df["Age"] = df["Age"].fillna(age)
df
```

```python
df.isnull().sum()
```

# Employee Count by Department

```python
plt.figure(figsize = [10,4])
sns.set_style(style = "whitegrid")
sns.countplot(data = df , x = "Department" ,hue = "Gender", palette = "viridis")
plt.title("Employee Count by Department",fontweight = "bold",fontsize = 20)
plt.xlabel("Department",fontweight = "bold",fontsize = 16)
plt.ylabel("Employee Count",fontweight = "bold",fontsize = 16)
plt.show()
```

# Employee Count bye Age & Gender

```python
gb = df.groupby(["Age","Gender"]).size().reset_index(name="Employee Count")

plt.figure(figsize = [10,4])
sns.lineplot(data = gb , x = "Age" , y = "Employee Count" ,hue = "Gender",marker = "s")
plt.title("Employee Count bye Age & Gender",fontweight = "bold",fontsize = 20)
plt.xlabel("Age",fontweight = "bold",fontsize = 16)
plt.ylabel("Employee Count",fontweight = "bold",fontsize = 16)
plt.show()
```

# Strength by Marital Status

```python
ms = df.groupby("MaritalStatus").size()
print (ms)

#   MaritalStatus
#   [ms.index ]                  [ms.values]
#   Divorced                     327
#   Married                      673
#   Single                       470


plt.pie(ms.values ,labels = ms.index , autopct = "%.1f%%", shadow = True,radius = 1.2)
plt.title("Strength by marital status")
plt.show()
```

# Average Salary By Department 

```python
# renamed column MonthlyRate to Salary
df = df.rename(columns = {"MonthlyRate":"Salary"})
# print(df["Salary"])

gp1 = df.groupby(["Department"]).agg({"Salary":"mean"})
print(gp1)

sns.heatmap(data = gp1 , annot = True)
plt.title("Average Salary By department",fontweight = "bold")
plt.show()
```

# Average Salary By Department and Gender

```python
gp2 = df.groupby(["Department","Gender"]).agg({"Salary":"mean"})

plt.figure(figsize = [8,4])
sns.barplot(data = gp2, x = "Department" , y = "Salary", hue = "Gender",palette = "viridis")
plt.legend(bbox_to_anchor = (1,1))
plt.title("Average Salary By Department and Gender",fontweight = "bold")
plt.xlabel("Department",fontweight = "bold")
plt.ylabel("Average Salary",fontweight = "bold")
plt.show()
```

# Employee Count by Age

```python
plt.figure(figsize = [10,4])
sns.histplot(data = df , x = "Age" ,bins = 20,hue = "Gender", palette = "spring")
plt.title("Employee Count by Age",fontweight = "bold")
plt.xlabel("Age",fontweight = "bold")
plt.ylabel("Employee Count",fontweight = "bold")
plt.show()


# sns.countplot(data = df , x = "Age")
# plt.show()
```
